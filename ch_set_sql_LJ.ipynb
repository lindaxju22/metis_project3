### Problem 5

Which players have the longest career? Assume that the `debut` and `finalGame` columns comprise the start and end, respectively, of a player's career. *Hint:* Create a new table from 'Master.csv'. Also note that strings can be converted to dates using the [`DATE`](https://wiki.postgresql.org/wiki/Working_with_Dates_and_Times_in_PostgreSQL#WORKING_with_DATETIME.2C_DATE.2C_and_INTERVAL_VALUES) function and can then be subtracted from each other yielding their difference in days.

*Answer: altroni01 had the longest careers*

```python
query="""
  SELECT playerID, (finalGame - debut) as career_length
  FROM master
  WHERE NOT (debut IS NULL OR finalGame IS NULL)
  ORDER BY career_length DESC;
"""
cursor.execute(query)
cursor.fetchall()
```

### Problem 6

What is the distribution of debut months? *Hint:* Look at the `DATE` and [`EXTRACT`](https://www.postgresql.org/docs/current/static/functions-datetime.html#FUNCTIONS-DATETIME-EXTRACT) functions.

```python
query="""
  SELECT EXTRACT(MONTH FROM debut) as debut_month, COUNT(EXTRACT(MONTH FROM debut)) as count
  FROM master
  WHERE NOT (debut IS NULL)
  GROUP BY debut_month
  ORDER BY debut_month;
"""
cursor.execute(query)
cursor.fetchall()
```

### Problem 7

What is the effect of table join order on mean salary for the players listed in the main (master) table? *Hint:* Perform two different queries, one that joins on playerID in the salary table and other that joins on the same column in the master table. You will have to use left joins for each since right joins are not currently supported with SQLalchemy.

*Answer: Left join on the salary table will show average salaries for all players in the salary table. Left join on the master table will try to show average salaries for all players in the master table. However, not all of them have data in the salary table, therefore, they appear as 'None'.*

```python
query="""
  SELECT master.playerID, AVG(salary)
  FROM salaries
  LEFT JOIN master ON salaries.playerID = master.playerID
  GROUP BY master.playerID
  ORDER BY master.playerID;
"""
cursor.execute(query)
cursor.fetchall()
```

### Problem 1

Which team scored the most points when playing at home?

*Answer: Real Madrid CF scored 10 points at home*

```python
query="""
  SELECT home_team_api_id, team_long_name, MAX(home_team_goal) as most_points_home
  FROM match
  LEFT JOIN team ON match.home_team_api_id = team.team_api_id
  GROUP BY home_team_api_id
  ORDER BY most_points_home DESC;
"""
cursor_soccer.execute(query)
cursor_soccer.fetchall()
```

### Problem 5

What was the median tie score? Use the value determined in the previous question for the number of tie games. *Hint:* PostgreSQL does not have a median function. Instead, think about the steps required to calculate a median and use the [`WITH`](https://www.postgresql.org/docs/8.4/static/queries-with.html) command to store stepwise results as a table and then operate on these results.

*Answer: The median tie score was 1*

```python
query="""
    SELECT ROUND(AVG(home_team_goal), 2) AS median_item_count  
    FROM (  
      SELECT home_team_goal
      FROM (
        SELECT home_team_goal,
               COUNT(*) OVER() AS row_count,
               ROW_NUMBER() OVER(ORDER BY home_team_goal) AS row_number
        FROM match
        WHERE home_team_goal = away_team_goal
      )
    WHERE row_number IN ((row_count + 1)/2, (row_count + 2)/2)  
    );
"""
cursor_soccer.execute(query)
cursor_soccer.fetchall()
```
